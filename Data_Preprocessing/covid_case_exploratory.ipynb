```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Resources/Data/covid_case_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# death count plot
plt.figure(figsize=(10,8))
sns.countplot(y='death', data=df, color='red', saturation=0.5)
```

```python
# race/ethnicity count plot
plt.figure(figsize=(10,8))
sns.countplot(y='race_ethnicity', data=df, color='cyan', saturation=0.3)
```

```python
# compare death counts across race_ethnicity
sns.catplot(
    y='race_ethnicity',
    col='death',
    data=df,
    kind='count',
    color='cyan',
    saturation=0.3
);
```

```python
# age group count plot
plt.figure(figsize=(10,8))
sns.countplot(y='age_group', data=df, color='limegreen', saturation=0.8,
              order=['0 - 9 Years', '10 - 19 Years', '20 - 29 Years', '30 - 39 Years',
                     '40 - 49 Years', '50 - 59 Years', '60 - 69 Years', '70 - 79 Years',
                     '80+ Years'])
```

```python
# compare death counts across age group
sns.catplot(
    y='age_group',
    col='death',
    data=df,
    kind='count',
    color='limegreen',
    saturation=0.8,
    order=['0 - 9 Years', '10 - 19 Years', '20 - 29 Years', '30 - 39 Years',
           '40 - 49 Years', '50 - 59 Years', '60 - 69 Years', '70 - 79 Years',
           '80+ Years']
);
```

```python
# sex count plot
plt.figure(figsize=(10,8))
sns.countplot(y='sex', data=df, color='orange', saturation=0.9)
```

```python
# compare death counts across sex
sns.catplot(
    y='sex',
    col='death',
    data=df,
    kind='count',
    color='orange',
    saturation=0.9
);
```

```python
# plot case count over time
date_df = pd.to_datetime(df['date'])
plt.figure(figsize=(15, 5))
sns.lineplot(
    x=date_df.value_counts().index,
    y=date_df.value_counts().values
)
```
